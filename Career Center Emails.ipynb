```python
import urllib3.request
from bs4 import BeautifulSoup
import csv
import codecs
import certifi
```

```python
def remove_non_ascii_1(text):
    return ''.join([i if ord(i) < 128 else ' ' for i in text])
```

```python
site = 'http://www.nacelink.com/nl_schools_list.php'

http_pool = urllib3.connection_from_url(site)
page = http_pool.urlopen('GET',site)

soup = BeautifulSoup(page.data)
```

```python
# print(soup.prettify())
```

```python
urls = []
schools = []

for link in soup.find_all('a', attrs={'class':'st'}):
    url = link.get('href')
    text = link.findAll(text=True)
    urls.append(url)
    schools.append(text)
```

```python
schools = [item for sublist in schools for item in sublist]
```

```python
dict = {key:value for key, value in zip(schools, urls)}
```
